# Ownership distribution
Summarize how many names each owner controls.

```python
from xlm_ns_sdk import NameRecord, ownership_distribution
records = [
    NameRecord(owner='GA', registered_at=1, expires_at=10, grace_period_ends_at=12, resolver=None),
    NameRecord(owner='GA', registered_at=2, expires_at=11, grace_period_ends_at=13, resolver=None),
    NameRecord(owner='GB', registered_at=3, expires_at=12, grace_period_ends_at=14, resolver=None),
]
ownership_distribution(records)
```
